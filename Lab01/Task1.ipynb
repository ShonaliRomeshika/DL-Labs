```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random 100000x1 array from exponential distribution with scale parameter 1.0
exponential_array = np.random.exponential(scale=1.0, size=(100000, 1))

# Create histograms
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="Uniform (rand)")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="orange", label="Normal (randn)")
plt.hist(exponential_array, density=True, bins=100, histtype="step", color="green", label="Exponential")

# Adjust plot limits and labels
plt.axis([-1, 6, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of x and y values
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)

# Create a grid of x and y values
X, Y = np.meshgrid(x_range, y_range)

# Calculate the corresponding Z values for the function Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title of the plot
ax.set_title('3D Plot of Z = X**2 + Y**2')

# Show the plot
plt.show()
```

```python
# Pandas for managing datasets
import pandas as pd


# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns


!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')



# Display first 5 observations
df.head()


# Calculate Pearson correlation coefficients
pearson_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')

# Calculate Spearman correlation coefficients
spearman_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

# Create subplots for the heatmaps
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation')

# Plot Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Correlation')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
